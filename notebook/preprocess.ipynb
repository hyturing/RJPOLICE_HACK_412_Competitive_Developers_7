```python
import pandas as pd
```

```python
account_activity = pd.read_csv("Dataset/Data/Customer Profiles/account_activity.csv")
account_activity.head()
```

```python
customer_data = pd.read_csv("Dataset/Data/Customer Profiles/customer_data.csv")
customer_data.head()
```

```python
customer_suspicious_activity = pd.read_csv("Dataset/Data/Fraudulent Patterns/suspicious_activity.csv")
customer_suspicious_activity.head()
```

```python
customer_transaction_amt = pd.read_csv("Dataset/Data/Transaction Data/transaction_records.csv")
customer_transaction_amt.head()
```

```python
customer = pd.merge(account_activity, customer_data, on="CustomerID", how="outer")
customer
```

```python
customer = pd.merge(customer, customer_transaction_amt, on="CustomerID", how="outer")
customer.head()
```

```python
customer = pd.merge(customer, customer_suspicious_activity, on="CustomerID", how="outer")
customer.head()
```

```python
customer.to_csv("customer_info.csv", index=False)
```

```python
merchant_data = pd.read_csv("Dataset/Data/Merchant Information/merchant_data.csv")
merchant_data.head()
```

```python
transaction_category = pd.read_csv("Dataset/Data/Merchant Information/transaction_category_labels.csv")
transaction_category.head()
```

```python
transaction_amount = pd.read_csv("Dataset/Data/Transaction Amounts/amount_data.csv")
transaction_amount.head()
```

```python
transaction_anomaly = pd.read_csv("Dataset/Data/Transaction Amounts/anomaly_scores.csv")
transaction_anomaly.head()
```

```python
transaction = pd.merge(transaction_category, transaction_amount, on="TransactionID", how="outer")
transaction.head()
```

```python
transaction = pd.merge(transaction, transaction_anomaly, on="TransactionID", how="outer")
transaction.head()
```

```python
transaction_merchant = pd.read_csv("Dataset/Data/Transaction Data/transaction_metadata.csv")
transaction_merchant.head()
```

```python
transaction = pd.merge(transaction, transaction_merchant, on="TransactionID", how="outer")
transaction.head()
```

```python
merchant_data = pd.merge(merchant_data, transaction, on="MerchantID", how="outer")
merchant_data.head()
```

```python
merchant_data.to_csv("merchant_info.csv", index=False)
```
